```python
import pandas as pd
from datetime import datetime, timedelta
import plotly.express as px
```

```python
df=pd.read_csv("final_archive.csv")
df.columns=['Voltaje','Corriente','Potencia Activa','Potencia Aparente','Potencia Reactiva','Factor de Potencia']
df
```

### Tiempos
Debemos tener en cuenta que el proyecto fue iniciado a las 9:56 pm el 5 de marzo, y de acuerdo al código de arduino, se obtienen los datos cada 2 minutos 53 segundos, en ese orden de ideas, podemos saber exactamente en que momento se hicieron las mediciones para cada fila del df, y podemos generar una columna que describa lo anterior a tavés de un array:

```python
hora_inicial=datetime(year=2024, month=3, day=5, hour=21, minute=55, second=46)
aumento_segundos=172.8# Hace referencia al delay en el código de arduino
# Usamos una comprensión de listas que hace el aumento en segundos para cada lectura del sensor
momentos=[hora_inicial+timedelta(seconds=aumento_segundos*i) for i in range(1,df.shape[0]+1)]
df['Hora']=momentos
df
```

## Visualización de datos usando Plotly

```python
from plotly.subplots import make_subplots
fig=make_subplots(rows=2,cols=1)
fig.add_trace(px.line(data_frame=df,x='Hora',y='Voltaje',title='Voltaje a través del tiempo').data[0],row=1,col=1)
fig.add_trace(px.line(data_frame=df,x='Hora',y='Factor de Potencia',color_discrete_sequence=['red']).data[0],row=2,col=1)
fig.update_layout(title='FP y Voltaje',height=700,width=800)
fig.update_yaxes(title_text='Voltaje',row=1,col=1)
fig.update_yaxes(title_text='Factor de Potencia',row=2,col=1)
fig.update_xaxes(title_text='Hora',row=2,col=1)
fig.show()
```

Para esta visualización, podemos evidenciar diversas caídas de voltaje a diferentes horas del día (5:40 am, 9:40 am, 4:00 pm) que se pueden deber al funcionamiento interno del computador y la manera en la que consume voltaje de acuerdo a sus ciclos, o a la red misma.
Dichos picos no parecen mostrar afectaciones directas sobre el fp.

```python
fig2=px.line(data_frame=df,x='Hora',y=['Corriente','Factor de Potencia'],title='Corriente a través del tiempo')
fig2.update_yaxes(title_text='Corriente[A] y FP')
```

Podemos ver que a la 1:00 am del 6 de Marzo, la corriente que consume el computador bajó drásticamente de 800 mA a 600mA aprox. 

Esto está directamente relacionado con que para este momento, apagué la pantalla del PC para un menor consumo, sin embargo se puede ver que en realidad el fp de potencia era menos eficiente de esta manera, mostrando una disminución y bajo desempeño durante todo el tiempo que pasó lo anterior.
Luego, a las 10 pm del 6 de Marzo, se volvió a encender la pantalla, mostrando una mayor corriente consumida.
El factor de potencia se estabilizó a 0.77 para este momento.

## Potencias vistas en gráfica
Pasamos a graficar el comportamiento de los distintos tipos de potencias a través del tiempo de monitoreo:


```python
fig3=(px.line(data_frame=df,x='Hora',y=['Potencia Activa','Potencia Aparente','Potencia Reactiva','Factor de Potencia']))
fig_final=make_subplots(rows=4,cols=1)
fig_final.add_trace(fig3.data[0],row=1,col=1)
fig_final.add_trace(fig3.data[1],row=2,col=1)
fig_final.add_trace(fig3.data[2],row=3,col=1)
fig_final.add_trace(fig3.data[3],row=4,col=1)
fig_final.update_layout(title='Potencias[W] y FP vs Tiempo',height=1200,width=900)
fig_final.update_yaxes(title_text='Potencia Activa',row=1,col=1)
fig_final.update_yaxes(title_text='Potencia Aparente',row=2,col=1)
fig_final.update_yaxes(title_text='Potencia Reactiva',row=3,col=1)
fig_final.update_yaxes(title_text='Factor de Potencia',row=4,col=1)
fig_final.update_xaxes(title_text='Hora')
```

# Análisis calculos realizados
Para este proyecto, podemos relacionar gracias a estas gráficas la relación directa entre la potencia activa (obtenida directamente desde el sensor) y la potencia aparente (calculada de forma teórica multiplicando el voltaje por la corriente)
La relación entre ambas se ve representada en el factor de potencia FP, que a medida que es más cercano a 1, significa que los valores anteriormente descritos son más parecidos.
Podemos ver que en realidad, este factor no fue generalmente cercano a 1, ya que la potencia teórica calculada de acuerdo a el voltaje y corriente consumidos, no es acorde a la real potencia consumida por el computador que es detectada en tiempo real por el sensor.
### Conclusiones
- Podemos decir que el sensor tiene en cuenta otros factores para entregar un valor preciso de la potencia activa, más allá de simplemente multiplicar voltaje por corriente.
- Se nota un factor de potencia cuestionable en horas de no uso del computador, ya que en estos momentos, el FP se vió aproximado a valores entre 0.65 y 0.75
- La potencia reactiva también se ve influenciada por momentos en los que se usó el equipo y no, lo cual tiene sentido al entender que dicha potencia se obtiene del triángulo de potencias visto en clase.
